```python
def getProductOfferslData(db, api_password, upc_product_code, seller_email):
    sql = '''
        Select 
        max(tbl_products.id) as id, 
        max(tbl_products.upc_product_code) as  upc_product_code, 
        max(tbl_products.seller_email) as seller_email, 
        max(tbl_products.catalog_id) as catalog_id,
        max(title) as title, 
        max(description) as description, 
        max(category) as category,
        max(sub_category) as sub_category, 
        max(parameter1) as parameter1, 
        max(parameter2) as parameter2, 
        max(parameter3) as parameter3, 
        max(parameter4) as parameter4, 
        max(parameter5) as parameter5, 
        max(bulk_or_individual) as bulk_or_individual, 
        max(product_status) as product_status, 
        max(buyer_offers.created_at) as created_at, 
        max(buyer_offers.updated_at) as updated_at, 
        MIN(buyer_offer_price) AS min_buyer_offer_price, 
        MAX(buyer_offer_price) AS max_buyer_offer_price, 
        AVG(averagePrice) AS avg_buyer_offer_price, 
        count(*) AS total_offers, 
        MIN(buyer_highest_offer_price) AS min_buyer_highest_offer_price, 
        MAX(buyer_highest_offer_price) AS max_buyer_highest_offer_price, 
        AVG(buyer_highest_offer_price) AS avg_buyer_highest_offer_price
        
        
        from 
        
        (select 
        *, 
        (buyer_offer_price + buyer_highest_offer_price)/2 as averagePrice
        
        from tbl_buyer_offers) buyer_offers
        
        join tbl_products on (tbl_products.upc_product_code = buyer_offers.upc_product_code)
        join tbl_service_requests on (tbl_service_requests.id = buyer_offers.id)
        
        where 
        tbl_service_requests.api_password = '{}' and
        tbl_products.upc_product_code = '{}' and 
        tbl_products.seller_email = '{}' 
        group by tbl_products.upc_product_code;
        '''.format(api_password, upc_product_code, seller_email)

    returnObject = {}
    results = pd.read_sql_query(sql, db.myDB)
    returnObject['success'] = 'true'
    returnObject['data'] = [];

    for i in range(0, len(results)):
        lineObject = {}
        lineObject['id'] = int(results.iloc[i]['id'])
        lineObject['upc_product_code'] = str(results.iloc[i]['upc_product_code'])
        lineObject['title'] = str(results.iloc[i]['title'])
        lineObject['description'] = str(results.iloc[i]['description'])
        lineObject['category'] = str(results.iloc[i]['category'])
        lineObject['sub_category'] = str(results.iloc[i]['sub_category'])		
        lineObject['parameter1'] = 	str(results.iloc[i]['parameter1'])
        lineObject['parameter2'] = 	str(results.iloc[i]['parameter2'])
        lineObject['parameter3'] = 	str(results.iloc[i]['parameter3'])
        lineObject['parameter4'] = 	str(results.iloc[i]['parameter4'])
        lineObject['parameter5'] = 	str(results.iloc[i]['parameter5'])
        lineObject['bulk_or_individual'] = str(results.iloc[i]['bulk_or_individual'])
        lineObject['product_status'] = str(results.iloc[i]['product_status'])
        lineObject['seller_email'] = str(results.iloc[i]['seller_email'])
        lineObject['catalog_id'] = int(results.iloc[i]['catalog_id'])
        lineObject['created_at'] = str(results.iloc[i]['created_at'])
        lineObject['updated_at'] = str(results.iloc[i]['updated_at'])
        lineObject['min_buyer_offer_price'] = float(results.iloc[i]['min_buyer_offer_price'])
        lineObject['max_buyer_offer_price'] = float(results.iloc[i]['max_buyer_offer_price'])
        lineObject['avg_buyer_offer_price'] = float(results.iloc[i]['avg_buyer_offer_price'])
        lineObject['min_buyer_highest_offer_price'] = float(results.iloc[i]['min_buyer_highest_offer_price'])
        lineObject['max_buyer_highest_offer_price'] = float(results.iloc[i]['max_buyer_highest_offer_price'])
        lineObject['avg_buyer_highest_offer_price'] = float(results.iloc[i]['avg_buyer_highest_offer_price'])
        returnObject['data'].append(lineObject)

    return returnObject
```
